```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv ("polynomial.csv" , delimiter = ";")
data
```

```python
polynomial_regression = PolynomialFeatures(degree=4) # degree polynomial denklemde ne kadar uzayacağını gösterir, ne kadar büyükse o kadar iyi
```

```python
plt.scatter (data["deneyim"],data["maas"]) # plt.scatter noktalı bağımsız graph gösterir

plt.xlabel("Experience (Year)")
plt.ylabel("Salary")

plt.show()
```

```python
# Linear regression çizimin yanlışlığının kanıtı

reg = LinearRegression()
reg.fit (data [ ["deneyim"] ] , data["maas"])

plt.scatter (data["deneyim"],data["maas"])
plt.plot (data["deneyim"],reg.predict(data[["deneyim"]]),color="red" , label = "Linear Regression" )
plt.legend()
plt.show()
```

```python
#Polynomial çizim

polynomial_regression = PolynomialFeatures(degree = 4)

x_polynomial = polynomial_regression.fit_transform (data[["deneyim"]]) # deneyime göre maaş hesaplayacak, deneyimi fit et

reg = LinearRegression()
reg.fit (x_polynomial , data["maas"]) # deneyimin polynomial olarak eğitilmiş halini ver, maaş aynı şekilde

reg.predict (polynomial_regression.fit_transform ([[4.5]])) #4.5 yıllık tecrübeli birinin maaş tahmini

```

```python
plt.plot(data["deneyim"] , reg.predict(x_polynomial) , color = "purple" , label = "Polynomial Regression") #predict value
plt.scatter(data["deneyim"],data["maas"]) #gerçek değerlerin grafiği
plt.xlabel("Experience (Year)")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
